K-Nearest Neighbors (KNN) imputation can handle both categorical and numerical data, but it requires some preprocessing. Since KNN relies on distances, categorical features must be encoded properly.

continent	✅ One-Hot Encoding	Small, unordered categories (prevents false relationships).
location	✅ Label Encoding	Unique country names (avoids high cardinality).
date	✅ Convert to DateTime & Extract Features

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df.drop(columns=["code"], inplace=True)

df["date"] = pd.to_datetime(df["date"])
df["day_of_year"] = df["date"].dt.dayofyear
df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.weekday

# Apply Label Encoding to location (since iso_code is dropped)
le_location = LabelEncoder()
df["country"] = le_location.fit_transform(df["country"])

# Apply One-Hot Encoding to continent
df = pd.get_dummies(df, columns=["continent"])
```

```python
import pandas as pd

df = pd.read_excel('/content/alcoholdata.xlsx',skiprows=2)


print("Columns in dataset:\n", df.columns.tolist())
```

```python
df_selected = df[[
    'SpatialDimValueCode',  # Country Code
    'Location',             # Country Name
    'Period',               # Year
    'Dim1',                 # Alcohol Type (e.g. Beer, Wine, All types)
    'Value',                # Alcohol consumption in litres
]]
```

```python
df_selected.columns = ['CountryCode', 'Country', 'Year', 'AlcoholType', 'AlcoholConsumption']
```

```python
df_selected.head(10)
```

we are analyzing alcohol consumption in general (e.g., for correlation with COVID-19 deaths), then using only:

➤ AlcoholType == "All types"
is the best and most reliable choice because:



```python
# Filter to include only 'All types'
df_all_types = df_selected[df_selected['AlcoholType'] == 'All types']

# Drop 'AlcoholType' column since it's now redundant
df_clean = df_all_types.drop(columns='AlcoholType')

# Rename for clarity
df_clean = df_clean.rename(columns={
    'AlcoholConsumption': 'TotalAlcoholLiters'
})


df_clean.head(21)
```

calculating the average alcohol consumption from 2000 to 2019 per country will give you a solid metric for how “alcoholic” a country generally is.

```python
# Filter years between 2000 and 2019 (inclusive)
df_filtered = df_clean[(df_clean['Year'] >= 2000) & (df_clean['Year'] <= 2020)]

# Group by Country and calculate mean alcohol consumption
df_avg = df_filtered.groupby(['CountryCode', 'Country'])['TotalAlcoholLiters'].mean().reset_index()

# Rename the column for clarity
df_avg = df_avg.rename(columns={'TotalAlcoholLiters': 'AvgAlcohol2000_2020'})

# Round to 2 decimal places (optional)
df_avg['AvgAlcohol2000_2020'] = df_avg['AvgAlcohol2000_2020'].round(5)

# Preview
df_avg.head(10)
```

```python
output_path = '/content/avg_alcohol_2000_2020.csv'
df_avg.to_csv(output_path, index=False)
```

**FOR SMOKING DATASET**

```python
import pandas as pd

smoking_df = pd.read_csv('/content/share-of-adults-who-smoke.csv')

smoking_df.head(40)
```

```python
# Filter years between 2000 and 2020 (inclusive)
smoking_filtered = smoking_df[(smoking_df['Year'] >= 2000) & (smoking_df['Year'] <= 2020)]

# Group by country code and name, calculate average prevalence
smoking_avg = smoking_filtered.groupby(['Code', 'Entity'])['Prevalence of current tobacco use (% of adults)'].mean().reset_index()

# Rename columns for consistency
smoking_avg = smoking_avg.rename(columns={
    'Code': 'CountryCode',
    'Entity': 'Country',
    'Prevalence of current tobacco use (% of adults)': 'AverageSmokersAdults'
})

# Round to 2 decimals (optional)
smoking_avg['AverageSmokersAdults'] = smoking_avg['AverageSmokersAdults'].round(2)

# Preview
smoking_avg.head(10)
```

Burundi (BDI) has an AverageSmokers value of 15.79, which means:

On average, 15.79% of adults in Burundi were current tobacco users (smokers) from 2000 to 2020.

**FOR OBESE DATASET**

```python
# Load the obesity dataset
obese_df = pd.read_csv('/content/share-of-adults-defined-as-obese.csv')

# Preview first 5 rows
obese_df.head(65)
```

```python
unique_years = obese_df['Year'].unique()
unique_years.sort()

print("Available years in obesity dataset:")
print(unique_years)
```

```python
# Step 1: Rename using exact long column name
obese_df = obese_df.rename(columns={
    'Entity': 'Country',
    'Code': 'CountryCode',
    'Prevalence of obesity among adults, BMI >= 30 (crude estimate) (%) - Sex: both sexes - Age group: 18+  years': 'ObesityRate'
})
```

```python
obese_df.head()
```

```python
obese_filtered = obese_df[(obese_df['Year'] >= 2000) & (obese_df['Year'] <= 2016)]

# Step 3: Group and average
obese_avg = obese_filtered.groupby(['CountryCode', 'Country'])['ObesityRate'].mean().reset_index()

# Step 4: Round and preview
obese_avg['AverageObesity'] = obese_avg['ObesityRate'].round(2)
obese_avg = obese_avg.drop(columns=['ObesityRate'])

obese_avg.head()
```

Prevalence of obesity:-How common obesity is in a population (as a percentage).

Adults, BMI ≥ 30	Adults whose Body Mass Index (BMI) is 30 or above, which is the medical definition of obesity.

Crude estimate (%)	A basic, unadjusted average — not accounting for factors like age structure or urban vs rural differences. Value is given in percentage (%).

Sex: both sexes	The data includes both men and women, not separated by gender.

Age group: 18+ years	Only adults aged 18 and above are included in the calculation.

 Example:
If AverageObesity = 27.5 for a country:

On average, 27.5% of adults aged 18+ (both men and women) in that country had a BMI ≥ 30 between 2000–2016.

```python
df['date'] = pd.to_datetime(df['date'])

# Step 2: Group by 'date' and keep only complete rows
df_cleaned = (
    df.groupby('date')
      .apply(lambda group: group.dropna(how='any'))  # keep only rows with all values present
      .reset_index(drop=True)
)
```
